```python
#URL Links for all pages
```

```python
urls=[]

for var in range(1,32):
    url=['https://www.flipkart.com/search?q=mi+mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=%i'%(var)]
    urls.append(url[0])
```

```python
urls
```

```python
#Defining a function to display contents of a page
```

```python
def connect(ur):
    uClient=urlopen(ur)
    page_html=uClient.read()
    uClient.close()
    page_soup = soup(page_html, "html.parser")
    container=page_soup.find_all('div',{'class':'_2kHMtA'})
    
    return container
```

```python
con=connect(urls[0])
con
```

```python
print(soup.prettify(con[0]))
```

```python
#Product Name(Testing the defined function whether it is display product names from all the pages)
```

```python
for i in range(0,31):
    con=connect(urls[i])
    for container in con:
        p_name=container.findAll('div',{'class':'_4rR01T'})
        print(p_name[0].text)
```

```python
#Creating a function to find product information from all the pages and store it in a dictionary
```

```python
def dataframe(containers):
    data={'Product Name':[],'Sale Price':[],'Actual Price':[],'Ratings':[],'Number of Ratings & Reviews':[],'Offer':[]}
    
    for i in range(0,31):
        con=connect(urls[i])
        for container in con:
            #Product Name
            product_name=container.findAll('div',{'class':'_4rR01T'})
            p_name=product_name[0].text
            data['Product Name'].append(p_name)
            
            #Sale Price:
            sale_price=container.findAll('div',{'class':'_30jeq3 _1_WHN1'})
            s_price=sale_price[0].text
            data['Sale Price'].append(s_price)
            
            #Actual Price:
            try:
                actual_price=container.findAll('div',{'class':'_3I9_wc _27UcVY'})
                a_price=actual_price[0].text
                data['Actual Price'].append(a_price)
            except IndexError:
                data['Actual Price'].append('Same as Sale Price')
                
            #Ratings
            try:
                ratings=container.findAll('div',{'class':'_3LWZlK'})
                r=ratings[0].text
                data['Ratings'].append(r)
            except IndexError:
                data['Ratings'].append('No Ratings')
                
            #Number of ratings and Reviews
            try:
                rat_n_rev=container.findAll('span',{'class':'_2_R_DZ'})
                n_r_r=rat_n_rev[0].text
                data['Number of Ratings & Reviews'].append(n_r_r)
            except IndexError:
                data['Number of Ratings & Reviews'].append('No_Ratings')
            
            #Offer
            try:
                offer=container.findAll('div',{'class':'_3Ay6Sb'})
                o=offer[0].text
                data['Offer'].append(o)
            except IndexError:
                data['Offer'].append('No Offer')
                
        print('done',i)
    
    return data
```

```python
dataset=dataframe(urls)
```

```python
#Storing in the form of a Dataset
```

```python
df_mi_phones=pd.DataFrame(dataset)
df_mi_phones
```
