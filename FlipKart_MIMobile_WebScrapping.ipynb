```python
#Importing the required libraries
```

```python
import numpy as np
import pandas as pd
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
#Displaying the URL Link
```

```python
url='https://www.flipkart.com/search?q=mi%20mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off'
```

```python
uClient=urlopen(url)
page_html=uClient.read()
uClient.close()
```

```python
#Displaying the contents of a page
```

```python
page_soup=soup(page_html,'html.parser')
```

```python
page_soup
```

```python
#Displaying the required content we need to parse
```

```python
container=page_soup.find_all('div',{'class':'_2kHMtA'})
```

```python
container
```

```python
con=container[0]
```

```python
print(soup.prettify(con))
```

```python
#Product Name
```

```python
for i in container:
    p_name=i.findAll('div',{'class':'_4rR01T'})
    print(p_name[0].text)
```

```python
#Sale Price
```

```python
for i in container:
    s_price=i.findAll('div',{'class':'_30jeq3 _1_WHN1'})
    print(s_price[0].text)
```

```python
#Actual Price
```

```python
for i in container:
    a_price=i.findAll('div',{'class':'_3I9_wc _27UcVY'})
    print(a_price)
```

```python
#Ratings
```

```python
for i in container:
    ratings=i.findAll('div',{'class':'_3LWZlK'})
    print(ratings[0].text)
```

```python
#No. of Ratings and Reviews
```

```python
for i in container:
    n_ratings_reviews=i.findAll('span',{'class':'_2_R_DZ'})
    print(n_ratings_reviews[0].text)
```
